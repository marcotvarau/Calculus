```python
import numpy as np
import matplotlib.pyplot as plt

# Define os intervalos de x para cada trecho
x_linear_esquerda = np.linspace(-20, 0, 200)  # trecho para x < 0
x_parabola = np.linspace(0, 30, 300)           # trecho para 0 < x < 30
x_linear_direita = np.linspace(30, 50, 200)      # trecho para x > 30

# Calcula os valores de y para cada trecho
y_linear_esquerda = 0.8 * x_linear_esquerda
y_parabola = -0.04 * x_parabola**2 + 0.8 * x_parabola
y_linear_direita = -1.6 * x_linear_direita + 36

# Cria o grfico
plt.figure(figsize=(8, 6))

# Plota as retas (bordas) em vermelho
plt.plot(x_linear_esquerda, y_linear_esquerda, color='red', label='Reta (x < 0)')
plt.plot(x_linear_direita, y_linear_direita, color='red', label='Reta (x > 30)')

# Plota a parbola em azul
plt.plot(x_parabola, y_parabola, color='blue', label='Parbola (0 < x < 30)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Montanha Russa - Funo por Partes')
plt.legend()
plt.grid(True)
plt.show()
```
